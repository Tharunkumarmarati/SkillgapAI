# Tech Stack Counter

```python
programming_languages = ["python","java","c","c++","javascript"]
Databases = ["mysql","postgresql","mongodb"]
frameworks = ["django","flask","react","angular"]
pl_count = 0
db_count = 0
fw_count = 0
tools = input("Enter tools/ technologies:")
tools_list = tools.split(",")
for tool in tools_list:
    if tool in programming_languages:
        pl_count += 1
    elif tool in Databases:
        db_count += 1
    elif tool in frameworks:
        fw_count += 1
print("Summary:")
print("Programming Languages:",pl_count)
print("Databases :",db_count)
print("Frameworks:",fw_count)
```
