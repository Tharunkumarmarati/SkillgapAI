## 9. Compare resume text and job description text to find common skills.

```python
resume_text = """
I am a software developer with experience in Python, HTML.
I have good communication skills and can work effectively in a team.
"""
resume_text_lower = resume_text.lower()
job_description = """
We are looking for a candidate skilled in Python, SQL, CSS and Java.
Good communication and teamwork skills are required.
"""
job_description_lower = job_description.lower()

technical_skills = ["Python", "Java", "SQL", "HTML", "CSS"]
soft_skills = ["Communication", "Teamwork", "Leadership"]
```

```python
# Convert text to lowercase
resume_lower = resume_text.lower()
job_desc_lower = job_description.lower()

# Extract skills from resume
resume_skills = []
for skill in technical_skills + soft_skills:
    if skill.lower() in resume_lower:
        resume_skills.append(skill)

# Extract skills from job description
job_skills = []
for skill in technical_skills + soft_skills:
    if skill.lower() in job_desc_lower:
        job_skills.append(skill)

# Find common skills
common_skills = set(resume_skills).intersection(set(job_skills))

print("Resume Skills:", resume_skills)
print("Job Description Skills:", job_skills)
print("Common Skills:", common_skills)
```

## 13. Count how many times each skill appears in the resume text.

```python
skill_count = {}

for skill in technical_skills + soft_skills:
    count = resume_text_lower.count(skill.lower())
    if count > 0:
        skill_count[skill] = count

print("\nSkill Frequency:", skill_count)
```

```python
matched_skills = []

with open("Skills2.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        if row["skill"].lower() in resume_text_lower:
            matched_skills.append(row["skill"])

print("Case-Insensitive Matched Skills:", matched_skills)
```

## 15. Save the extracted skills into a new CSV file.

```python
with open("extracted_skills.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Skill", "Category"])

    for skill in extracted_tech_skills:
        writer.writerow([skill, "Technical"])

    for skill in extracted_soft_skills:
        writer.writerow([skill, "Soft"])

print("\nExtracted skills saved to extracted_skills.csv")
```
