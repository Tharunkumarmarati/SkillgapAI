# Experience Extractor

```python
input_sentence = input("Enter experience sentence:")
input_sentence_list = input_sentence.split()
exp = ""
for i in input_sentence_list:
    if i.lower() == "experience":
        for j in input_sentence_list:
            if j.isdigit():
                exp = j
                print("Experience:",exp)
                break
    else:
        print("No experience detected")
```
