# Skill Frequency Checker

```python
resume_text = "I am a highly motivated software engineer with experience in Python, Java, and SQL. I have worked on multiple data analysis projects using Python and Pandas. I also have hands-on experience in web development with HTML, CSS, and JavaScript. My recent project involved building an automated data processing system using machine learning techniques."
for i in ["-",",","."]:
    resume_text = resume_text.replace(i,"")
User_skill = input("Enter a Skill:")
count = 0
resume_words = resume_text.split(" ")
for i in resume_words:
    if i.lower() == User_skill.lower():
        count += 1
print("count of User_skill:",count)
```
