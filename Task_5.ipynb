# Simple Resume Category Classifier

```python
data_ml = ["python","machine learning","deep learning","pandas"]
wed_dev = ["html","css","javascript","react","angular"]
soft_dev = ["java","c","c++","git","sql"]
skills = input("Enter skills:").lower()
skills_list = skills.split(",")
category = ""
for skill in skills_list:
    if skill in data_ml:
        category = "Data/ML"
        break
    elif skill in web_dev:
        category = "Web Development"
        break
    elif skill in soft_dev:
        category = "Software Development"
        break
    else:
        category = "Other"
print("Predicted Category:",category)
```
