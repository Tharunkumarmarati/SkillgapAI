```python
import json
```

## 1. Convert the JSON string {'name': 'Alex', 'age': 22, 'city': 'Chennai'} into a Python dictionary.

```python
json_str = '{"name": "Alex", "age": 22, "city": "Chennai"}'
pyt_dict = json.loads(json_str)
print("Python Dictionary:",pyt_dict)
print(type(pyt_dict))
```

## 2. Convert the dictionary {'id': 101, 'product': 'Laptop', 'price': 55000} into a JSON string.

```python
dict1 = {'id': 101, 'product': 'Laptop', 'price': 55000}
pyt_json_str = json.dumps(dict1)
print("Python dictionary to json str:",pyt_json_str)
print(type(pyt_json_str))
```

## 6. Update the dictionary {'name': 'Raj', 'age': 45} by changing age to 50 and convert it to JSON.

```python
py_dict = {'name':'Raj','age' : 45}
py_dict['age'] = 50
dict_to_json = json.dumps(py_dict,indent = 4)
print(dict_to_json)
print(type(dict_to_json))
```

## 8. From JSON {'users': [{'id':1,'name':'Alex'}, {'id':2,'name':'Mia'}]}, print all user names.

```python
json_file = '{"users": [{"id":1,"name":"Alex"}, {"id":2,"name":"Mia"}]}'
py_dicti = json.loads(json_file)
for i in py_dicti['users']:
    print(i['name'])
```

## 9. Add a new key 'country': 'India' to {'name': 'Sam', 'age': 28} and convert back to JSON.

```python
dict1 = {"name":"Sam","age":28}
dict1['country'] = 'India'
dict1_json = json.dumps(dict1,indent = 4)
print(dict1_json)
```

## 10. Create items.json with [{'item':'Pen','price':10}, {'item':'Book','price':50}, {'item':'Bag','price':700}]and print total cost.

```python
data = [{'item':'Pen','price':10}, {'item':'Book','price':50}, {'item':'Bag','price':700}]
```

```python
with open(r"C:\Users\tarun\Infosys_springboard\csv_json\items.json",'w') as file:
    json.dump(data,file,indent = 4)
```

```python
with open(r"C:\Users\tarun\Infosys_springboard\csv_json\items.json",'r') as file:
    items_data = json.load(file)
total_cost = 0
for item in items_data:
    total_cost += int(item['price'])
print("Total_cost:",total_cost)
```
