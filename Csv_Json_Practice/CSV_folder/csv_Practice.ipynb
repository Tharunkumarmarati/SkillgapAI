## 3. From products.csv, print only the first column (product names).

## 4. Read a CSV using DictReader and print all key-value pairs for each row.

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/Products.csv',mode = 'r') as file:
    read_file = csv.DictReader(file)
    for row in read_file:
        for key,value in row.items():
            print(f'{key}: {value}')
        print('-'*15)
```

## 5. Add a new row to marks.csv with columns name, subject, marks.

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/marks.csv',mode = 'a',newline = "") as file:
    marks_file = csv.DictWriter(file,fieldnames = ["name","subject","marks"])
    marks_file.writerow({"name":"Vinay","subject":"Hindi","marks":95})
```

## 6. Count how many rows exist in data.csv (excluding the header).

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/data.csv',mode = 'r') as file:
    data_reader = csv.DictReader(file)
    no_of_rows = 0
    for row in data_reader:
         no_of_rows += 1
print("Number of rows in data.csv:",no_of_rows)
```

## 7. Read sales.csv and calculate total sales (item, price, quantity).

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/sales.csv',mode = 'r') as file:
    sales_data = csv.DictReader(file)
    total_sales = 0
    for row in sales_data:
        total_sales += int(row['price'])
print("Total sales:",total_sales)
```

## 9. Filter and print rows from people.csv where age > 30

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/people.csv',mode = 'r') as file:
    people = csv.DictReader(file)
    for row in people:
        if int(row['age']) > 30:
            print(row)
```

## 10. Copy all contents from source.csv to copy.csv.

```python
with open(r'C:/Users/tarun/Infosys_springboard/csv_json/people.csv',mode = 'r') as src, open(r'C:/Users/tarun/Infosys_springboard/csv_json/people_copy.csv',mode = 'w',newline = "") as dest:
    ppl = csv.reader(src)
    ppl_copy = csv.writer(dest)
    ppl_copy.writerows(ppl)
```
