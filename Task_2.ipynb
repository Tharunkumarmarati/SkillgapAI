# Resume word Counter

```python
resume_paragraph = "I am a dedicated and results-driven Computer Science Engineering student with strong skills in Java, Python, HTML, CSS, and JavaScript. Passionate about building real-world software applications, I have worked on several academic and personal projects including an Online Examination System, a Secured Grading Examination System, and multiple Python-based games. With a strong interest in Python Full Stack Development, I strive to continuously improve my technical skills and adapt to new technologies. I am eager to contribute to a dynamic organization where I can apply my knowledge, grow professionally, and create impactful solutions."
resume_paragraph = resume_paragraph.replace(".","")
resume_paragraph = resume_paragraph.replace(",","")
resume_paragraph_list = resume_paragraph.split(" ")
print("count =", len(resume_paragraph_list))
print("Unique_count =",len(set(resume_paragraph_list)))   
dict_ = {}
for word in resume_paragraph_list:
    if word in dict_:
        dict_[word] += 1
    else:
        dict_[word] = 1 
max_word = ""
max_count = 0
for word in dict_:
    if dict_[word] > max_count:
        max_count = dict_[word]
        max_word = word
print("Most Repeated word =",max_word)
```
