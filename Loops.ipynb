```python
# 1. Print numbers from 1 to 10 using a for loop.
print("1. Numbers from 1 to 10 using a for loop")
for i in range(1,11):
    print(i)
# 2. Print all even numbers from 1 to 20.
print("2. Even numbers from 1 to 20")
for i in range(1,21):
    if i % 2 == 0:
        print(i)
# 3. Print each character of the string: 'Python'.
print("3. Each character of the string: 'Python'")
for i in 'Python':
    print(i)
# 4. Using a while loop, print numbers from 5 down to 1.
print("4. Using a while loop, print numbers from 5 down to 1")
k = 5
while k > 0:
    print(k)
    k -= 1
# 5. Write a loop to find the sum of numbers from 1 to 50.
print("5. sum of numbers from 1 to 50")
total = 0
for i in range(1,51):
    total += i
print("Total:", total)
# 6. Print the multiplication table of 5 (from 5×1 to 5×10).
print("6.  multiplication table of 5 (from 5×1 to 5×10)")
for i in range(1,11):
    print("5 *",i," = ",5 * i) 
# 7. Find how many vowels are present in the string 'Programming'.
print("7. Find how many vowels are present in the string 'Programming'")
vowels = ['a','e','i','o','u']
count = 0
for i in "Programming":
    if i.lower() in vowels:
        count += 1
print("No.of vowels =",count)
# 8. Use a loop to reverse the string 'PythonLoops'.
que_8 = 'PythonLoops'
print("8. Reverse the string 'PythonLoops'")
for i in range(len(que_8) - 1, -1,-1):
    print(que_8[i])
# 9. Print numbers from 1–10, but skip 5 using continue.
print("9. Numbers from 1–10, but skip 5 using continue")
for i in range(1,11):
    if i == 5:
        continue
    print(i)
# 10. Print numbers from 1–20, but stop when number reaches 13 using break.
print("10. numbers from 1–20, but stop when number reaches 13 using break")
for i in range(1,21):
    if i == 13:
        break
    print(i)
# 11. Write a loop to check if a number is prime.
print("11. loop to check if a number is prime.")
num = 19
count_2 = 0
if num > 1:
    for i in range(2,int(num ** 0.5) + 1):
        if num % i == 0:
            count_2 += 1
    if count_2 == 0:
        print(num,"is Prime")
    else:
        print(num,"is not Prime")
else:
    print(num,"is not Prime")
# 12. Count how many times each character occurs in 'mississippi'.
print("12. Count of each character occurs in 'mississippi'")
count_pair = {}
que_12 = "mississippi";
for i in que_12:
    if i in count_pair:
        count_pair[i] += 1
    else:    
        count_pair[i] = 1
print(count_pair)
# 13. Using nested loops, print the pattern: *, **, ***, ****, *****.
print("13. pattern: *, **, ***, ****, *****")
for i in range(1,6):
    for j in range(i):
        print("*",end = '')
    print()
# 14. Write a loop to find the largest digit in the number 5847361.
print("14. largest digit in the number 5847361")
temp = 1
number = 5847361
while number > 0:
    n = number % 10
    number = number // 10
    if n > temp:
        temp = n
print("Largest digit :",temp)
# 15. Write a loop to print the pattern: *****, ****, ***, **, *.
print("15. pattern: *****, ****, ***, **, *.")
for i in range(5,-1,-1):
    print("*"*i)
```
